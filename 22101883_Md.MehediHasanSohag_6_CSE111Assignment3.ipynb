```python
#Task-1
class Calculator:
    def __init__(self):
        self.var1=None
        self.var2=None
        self.opt=None
        self.rst=None
        print("Let’s Calculate!")
    def add(self):
        self.rst=self.var1+self.var2
        print("Result:",self.rst)
    def subtract(self):
        self.rst=self.var1-self.var2
        print("Result:",self.rst)
    def multiply(self):
        self.rst=self.var1*self.var2
        print("Result:",self.rst)
    def divide(self):
        if(self.var2==0):
            print("Result: Math Error!!")
        else:
            self.rst=self.var1/self.var2
            print("Result:",self.rst)
cal=Calculator()
cal.var1=int(input())
cal.opt=input()
cal.var2=int(input())
print("Value 1:",cal.var1)
print("Operator:",cal.opt)
print("Value 2:",cal.var2)
if(cal.opt=="+"):
    cal.add()
elif(cal.opt=="-"):
    cal.subtract()
elif(cal.opt=="*"):
    cal.multiply()
elif(cal.opt=="/"):
    cal.divide()
```

```python
#Task-5
class Shape:
    def __init__(self,name,d1,d2):
        self.name=name
        self.dimen1=d1
        self.dimen2=d2
    def area(self):
        if(self.name=="Triangle"):
            area=.5*self.dimen1*self.dimen2
        elif(self.name=="Rhombus"):
            area=.5*self.dimen1*self.dimen2
        elif(self.name=="Square"):
            area=self.dimen1*self.dimen2
        elif(self.name=="Rectangle"):
            area=self.dimen1*self.dimen2
        else:
            area="Shape unknown"
        print("Area:",area)
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#Task-6
class Calculator:
    def __init__(self):
        print("Calculator is ready!")
        self.num1=None
        self.num2=None
        self.opt=None
        self.result = None
    def calculate(self,var1,var2,var3):
        self.num1=var1
        self.num2=var2
        self.opt=var3
        if (self.opt=="+"):
            self.result=self.num1+self.num2
        elif (self.opt=="-"):
            self.result=self.result-self.num2
        elif (self.opt=="*"):
            self.result=self.result*self.num2
        elif (self.opt=="/"):
            self.result=self.result/self.num2
        return self.result
    def showCalculation(self):
        print(str(self.num1)+str(self.opt)+str(self.num2)+"="+str(self.result))
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#Task-7
class Student:
    def __init__(self,var1,var2,var3,var4):
        self.name=var1
        self.ID=var2
        self.dept=var3
        self.cgpa=var4
        self.fcgpa=None
    def calculate_CGPA(self):
        total=0
        for i in self.cgpa:
            total+=i*3
        num=len(self.cgpa)*3
        self.fcgpa=total/num
    def print_details(self):
        print(f"Name: {self.name}, ID:{self.ID}")
        print("Department:",self.dept)
        print("CGPA:",self.fcgpa)
        if(self.fcgpa>3.80):
            print("Your academic standing is 'Highest Distinction'")
        elif(self.fcgpa>3.65):
            print("Your academic standing is 'High Distinction'")
        elif(self.fcgpa>3.50):
            print("Your academic standing is 'Distinction'")
        elif(self.fcgpa>2.00):
            print("Your academic standing is 'Satisfactory'")
        else:
            print("Sorry, you cannot graduate")
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================")
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
#Task-11
class Spotify:
    def __init__(self,list1):
        print("Welcome to Spotify!") 
        self.slist=list1
    def playing_number(self,num):
        if(num>len(self.slist)):
            value=str(num)+" number song not found. Your playlist has "+str(len(self.slist))+" songs only."
            return value
        elif(num<=len(self.slist)):
            print("##########################")
            print("Playing",num,"number song for you")
            value="Song name: "+str(self.slist[num-1])
            return value
    def add_to_playlist(self,song):
        self.slist.append(song)
user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```

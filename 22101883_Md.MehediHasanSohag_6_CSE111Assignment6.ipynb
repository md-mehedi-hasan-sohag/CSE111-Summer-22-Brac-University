```python
#Task-7
class Cat:
    Number_of_cats=0
    def __init__(self,clr,do):
        self.clr=clr
        self.do=do
        Cat.Number_of_cats+=1
    def changeColor(self,clr):
        self.clr=clr
    def printCat(self):
        print(f"{self.clr} cat is {self.do}")
    @classmethod
    def no_parameter(cls):
        clr="White"
        do="sitting"
        data = Cat(clr,do)
        return data
    @classmethod
    def first_parameter(cls,clr):
        clr=clr
        do="sitting"
        data = Cat(clr,do)
        return data
    @classmethod
    def second_parameter(cls,do):
        clr="Grey"
        do=do
        data = Cat(clr,do)
        return data
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task-8
import math
class Cylinder:
    radius=5
    height=18
    def __init__(self,var1,var2):
        self.radius=var1
        self.height=var2
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        print(f"Updated: radius={self.radius} and height={self.height}.")
        Cylinder.radius=var1
        Cylinder.height=var2
    @classmethod
    def swap(cls,val1,val2):
        data=Cylinder(val2,val1)
        return data
    @classmethod
    def changeFormat(cls,string):
        val1,val2=string.split("-")
        val1=float(val1)
        val2=float(val2)
        data=Cylinder(val1,val2)
        return data
    @staticmethod
    def area(var1,var2):
        sarea=2*math.pi*math.pow(var1,2)+2*math.pi*var1*var2
        print(f"Area:  {sarea}")
    @staticmethod
    def volume(var1,var2):
        total_volume=math.pi*math.pow(var1,2)*var2
        print(f"Volume:  {total_volume}")
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class SultansDine:
    total_branch=0
    total_sell=0
    total_info=[]
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {SultansDine.total_branch}")
        print(f"Total Sell: {SultansDine.total_sell} Taka")
        for i in SultansDine.total_info:
            print(f"Branch Name: {i.place},Branch Sell: {i.branch_sell} Taka ")
            print(f"Branch consists of total sell's: {i.sell_percent}% ")
            
    def __init__(self,place):
        self.place=place
        self.quantity=0
        self.branch_sell=0
        SultansDine.total_branch+=1
        self.sell_percent=0
        SultansDine.total_info.append(self)
    def sellQuantity(self,quantity):
        self.quantity=quantity
        if self.quantity < 10:
            self.branch_sell = quantity * 300
        elif 10 < self.quantity < 20:
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400
        SultansDine.total_sell+=self.branch_sell 
        if len(SultansDine.total_info)>1:
            for i in SultansDine.total_info:
                i.sell_percent=round((i.branch_sell/SultansDine.total_sell)*100,2)
        else:
            self.sell_percent = round((self.branch_sell / SultansDine.total_sell) * 100,2)   
    def branchInformation(self):
        print(f"Branch Name: {self.place}")
        print(f"Branch Sell: {self.branch_sell} Taka")
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()      
```

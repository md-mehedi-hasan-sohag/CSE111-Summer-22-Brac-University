```python
#Task-7
class Match:
    def __init__(self,var):
        print("5..4..3..2..1.. Play !!!")
        var=var.split("-")
        self.bat=var[0]
        self.bowl=var[1]
        self.run=0
        self.over=0
        self.wicket=0
    def add_run(self,run):
        self.run+=run
    def add_over(self,over):
        if(self.over+over>5):
            print(f"Warning! Cannot add 5 over/s (5 over match)")
        else:
            self.over+=over
    def add_wicket(self,wicket):
        self.wicket+=wicket
    def print_scoreboard(self):
        print(f"Batting Team: {self.bat}")
        print(f"Bowling Team: {self.bowl}")
        value="Total Runs: "+str(self.run)+" Wickets: "+str(self.wicket)+" Overs: "+str(self.over)
        return value
match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```
